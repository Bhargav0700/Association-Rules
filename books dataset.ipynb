```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
```

```python
books=pd.read_csv("book.csv")
books
```

# APRIORI 


```python
frequent_itemsets=apriori(books,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rule1=association_rules(frequent_itemsets,metric="confidence",min_threshold=0.6)
rule1
```

```python
rule1[rule1["confidence"]>0.6]
```

```python
rule1
rule1.sort_values("confidence",ascending=False)
```

```python
rule1["antecedents"].value_counts()
```

# LIFT

```python
rule2=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rule2
```

```python
rule2.sort_values("lift",ascending=False)
```

```python
plt.subplots(figsize=(12,5))

plt.subplot(1,3,1)
plt.hist(rule2["support"])
plt.subplot(1,3,2)
plt.hist(rule2["confidence"])
plt.subplot(1,3,3)
plt.hist(rule2["lift"])
```
